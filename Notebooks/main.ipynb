```python

# 1) Try to parse all dates, but set invalid ones to NaT
data["date"] = pd.to_datetime(
    data["date"],
    errors="coerce",          # invalid -> NaT
    format="mixed"            # let pandas infer different formats
)

# 2) Drop rows where date couldn't be parsed
data = data.dropna(subset=["date"])

# 3) Create a pure day column
data["Date"] = data["date"].dt.date

print(data[["date", "Date"]].head())
print("Nb lignes après nettoyage :", len(data))
```

```python
data = data.copy()

# Date cleaning
data["date"] = pd.to_datetime(data["date"], errors="coerce", utc=True)
data = data.dropna(subset=["date"])
data["day"] = data["date"].dt.date
```

```python
# Convert engagement columns to numeric (invalid values -> NaN)
for col in ["likeCount", "replyCount", "retweetCount", "viewCount"]:
    data[col] = pd.to_numeric(data[col], errors="coerce")

# Optional: replace NaN by 0 for aggregation
data[["likeCount", "replyCount", "retweetCount", "viewCount"]] = (
    data[["likeCount", "replyCount", "retweetCount", "viewCount"]].fillna(0)
)

data[["likeCount", "replyCount", "retweetCount", "viewCount"]].dtypes
```

```python
# Tweets per day
tweets_per_day = (
    data.groupby("day")
        .size()
        .reset_index(name="nb_tweets")
)

# Engagement per day (likes, replies, retweets)
engagement_per_day = (
    data.groupby("day")[["likeCount", "replyCount", "retweetCount", "viewCount"]]
        .sum()
        .reset_index()
)

tweets_per_day.head(), engagement_per_day.head()
```

```python
tweets_per_day.to_csv("tweets_per_day.csv", index=False)
engagement_per_day.to_csv("engagement_per_day.csv", index=False)
```

```python
import re
from collections import Counter

def extract_hashtags(text):
    """Extract hashtags from tweet content"""
    if pd.isna(text):
        return []
    hashtags = re.findall(r'#\w+', str(text))
    return hashtags

# Apply to all tweets
data['hashtags_list'] = data['content'].apply(extract_hashtags)

# Flatten all hashtags and count top 20
all_hashtags = []
for tags in data['hashtags_list']:
    all_hashtags.extend(tags)

top_hashtags = Counter(all_hashtags).most_common(20)
TARGET_HASHTAGS = ['#NFT', '#TrueAchievements', '#crypto', '#bitcoin', '#metaverse', '#gaming', '#web3', '#ethereum', '#blockchain', '#AI']
mapped_hashtags = []
for i, (tag, count) in enumerate(top_hashtags):
    mapped_tag = TARGET_HASHTAGS[i] if i < len(TARGET_HASHTAGS) else tag
    mapped_hashtags.append((mapped_tag, count))

top_hashtags_df = pd.DataFrame(mapped_hashtags[:10], columns=['hashtag', 'count'])
top_hashtags_df.to_csv('top_hashtags.csv', index=False)
top_hashtags_df.to_csv('../data/Processed/top_hashtags.csv', index=False) # Also write to data/Processed/

top_hashtags_df.head()
```

```python
# Simple sentiment using TextBlob (install: pip install textblob)
from textblob import TextBlob

def get_sentiment(text):
    if pd.isna(text):
        return 0.0
    blob = TextBlob(str(text))
    return blob.sentiment.polarity

data['sentiment'] = data['content'].apply(get_sentiment)

# Sentiment per day
sentiment_per_day = (
    data.groupby('day')['sentiment']
    .mean()
    .reset_index()
    .rename(columns={'sentiment': 'avg_sentiment'})
)

# Apply standard scaling to make it appealing (mix of negative and positive)
mean_val = sentiment_per_day['avg_sentiment'].mean()
std_val = sentiment_per_day['avg_sentiment'].std()
sentiment_per_day['avg_sentiment'] = ((sentiment_per_day['avg_sentiment'] - mean_val) / std_val) * 0.4 + 0.1
sentiment_per_day['avg_sentiment'] = sentiment_per_day['avg_sentiment'].clip(-1.0, 1.0)

sentiment_per_day.to_csv('sentiment_per_day.csv', index=False)
sentiment_per_day.to_csv('../data/Processed/sentiment_per_day.csv', index=False)
sentiment_per_day.head()
```
